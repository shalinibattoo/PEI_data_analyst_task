```python
amt = spark.sql(f'''select c.Country,sum(o.Amount) from customer_df c
join order_df o on c.Customer_ID = o.Customer_ID
group by c.Country
''')
```

```python
amt.show()
```

```python
order_det = spark.sql(f'''select Customer_ID, count(distinct Order_ID) as transactions,
count(Item) as items,
sum(Amount) as total_amt
from order_df
group by Customer_ID
''')
```

```python
order_det.show(300)
```

```python
spark.sql('''select * from order_df where Customer_ID = "78"''').show()
```

```python
product = spark.sql(f'''select Country, Item
from
(select *,
max(count) over (partition by country) as max_count from
(select c.Country,o.Item, count(*) as count from customer_df c
join order_df o on c.Customer_ID = o.Customer_ID
group by 1,2)
)
where count = max_count
''')
```

```python
product.show()
```

```python
p1 = spark.sql('''select *,
max(count) over (partition by country) as max_count from
(select c.Country,o.Item, count(*) as count from customer_df c
join order_df o on c.Customer_ID = o.Customer_ID
group by 1,2)
''')
```
